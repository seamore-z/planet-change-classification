```python
aspectTif = rxr.open_rasterio('/projectnb/modislc/users/sjstone/above/data/dem/Qikiqtaruk_Aspect.tif')
```

```python
landPoly = gpd.read_file(mainPath + 'data/landPoly/landPoly.shp')
```

```python
clipped = aspectTif.rio.clip(landPoly['geometry'])
```

```python
aspectArray = clipped.values[0]
```

```python
aspectArray[aspectArray == -9999] = np.nan
```

```python
aspectArrayFlat = aspectArray.flatten()
aspectArrayFlat = aspectArrayFlat[~np.isnan(aspectArrayFlat)]
```

```python
plt.hist(aspectArrayFlat)
plt.title('Aspect Across Island')
plt.xlabel('Aspect')
plt.show()
```

```python
slopeTif = rxr.open_rasterio('/projectnb/modislc/users/sjstone/above/data/dem/Qikiqtaruk_Slope.tif')
slopeTifClipped = slopeTif.rio.clip(landPoly['geometry'])
```

```python
slopeArray = slopeTifClipped.values[0]
slopeArray[slopeArray == -9999] = np.nan
```

```python
slopeArrayFlat = slopeArray.flatten()
slopeArrayFlat = slopeArrayFlat[~np.isnan(slopeArrayFlat)]
```

```python
plt.hist(slopeArrayFlat, range = (0, 40))
plt.title('Slope Across Island')
plt.xlabel('Slope')
plt.show()
```
